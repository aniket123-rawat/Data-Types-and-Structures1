Ques 1 : What is Simple Linear Regression (SLR)? Explain its purpose.

Answer : Simple Linear Regression (SLR) is a statistical method that models the relationship between a single independent (predictor) variable and a single dependent (outcome) variable by fitting a straight line to the data. Its purpose is to understand how one variable changes in relation to another, to make predictions about the dependent variable based on the independent variable, and to assess the strength of the relationship between the two variables.

 SLR is used to find the best-fitting straight line through a set of data points, where each point represents a pair of values for two continuous variables.

 Two variables: It involves two variables: one independent (x) and one dependent (y).

 A linear relationship: It assumes the relationship between the variables can be described by a linear equation of the form \(y=\beta _{0}+\beta _{1}x+\epsilon \), where \(\beta _{0}\) is the intercept, \(\beta _{1}\) is the slope, and \(\epsilon \) is the error term.

 Best-fitting line: The goal is to find the values for the intercept (\(\beta _{0}\)) and slope (\(\beta _{1}\)) that minimize the sum of the squared differences between the observed and predicted values of the dependent variable.

 - To understand relationships: SLR provides a quantitative understanding of how a change in the independent variable is associated with a change in the dependent variable.
  For example, it can show how many points a test score (dependent variable) is expected to increase for each additional hour of studying (independent variable).
  
  - To make predictions: Once the line is established, you can use the equation to predict the value of the dependent variable for a new or unobserved value of the independent variable.
  
  - To quantify the strength of the relationship: The model helps determine how strongly the independent variable influences the dependent variable. This can be measured by metrics like the R-squared value, which indicates the proportion of the variation in the dependent variable that is predictable from the independent variable.
  
  - To serve as a foundation: It is a fundamental concept that forms the basis for more complex machine learning algorithms

Ques 2 : What are the key assumptions of Simple Linear Regression?

Answer : The main assumptions of simple linear regression are that the relationship between the independent and dependent variables is linear, the errors (residuals) are independent, the errors have constant variance (homoscedasticity), and the errors are normally distributed. For multiple linear regression, there is an additional assumption of no multicollinearity among predictors.

1. Linearity

What it means: The relationship between the independent variable (x) and the dependent variable (y) is a straight line.

How to check: Use a scatterplot to visualize the data and look for a linear trend.

2. Independence of errors

What it means: The residuals are not correlated with each other. One error does not affect another. This is particularly important for time-series data where errors might be correlated.

How to check: Use the Durbin-Watson test or examine a plot of residuals against the order of observations.

3. Homoscedasticity

What it means: The variance of the residuals is constant across all levels of the independent variable. In simpler terms, the spread of the residuals is roughly the same for all values of x.

How to check: Create a plot of residuals against the predicted values. The points should be randomly scattered with no discernible pattern.

4. Normality of errors

What it means: The residuals are normally distributed, meaning they follow a bell-shaped curve.

How to check: Use a Q-Q plot or normality tests like the Shapiro-Wilk test.

5. No multicollinearity

What it means: This is primarily for multiple linear regression and assumes that the independent variables are not highly correlated with each other.

How to check: Analyze the correlation matrix of the independent variables. If the correlation is too high, multicollinearity is an issue

Ques 3 : Write the mathematical equation for a simple linear regression model and
explain each term.

 Answer : The mathematical equation for a simple linear regression model is:

$$y = \beta_0 + \beta_1 x + \epsilon$$

Where:

- \(y\): This is the **dependent variable** (also known as the response or outcome variable). It's the variable you are trying to predict or explain.
- \(x\): This is the **independent variable** (also known as the predictor or explanatory variable). It's the variable you are using to predict \(y\).
- \(\beta_0\): This is the **y-intercept**. It represents the predicted value of \(y\) when the value of \(x\) is 0. It's the point where the regression line crosses the y-axis.
- \(\beta_1\): This is the **slope** of the regression line. It represents the change in \(y\) for every one-unit increase in \(x\). It indicates the direction and strength of the linear relationship between \(x\) and \(y\).
- \(\epsilon\): This is the **error term** (also known as the residual). It represents the random error or the part of \(y\) that cannot be explained by the linear relationship with \(x\). It accounts for unmeasured factors and random variability.

Ques 5 :  What is the method of least squares in linear regression?

Answer : The method of least squares in linear regression finds the "line of best fit" for a set of data points by minimizing the sum of the squared differences between the observed values and the values predicted by the line.

This method is used to create a predictive model that best represents the relationship between variables, making it a fundamental technique in statistics, machine learning, and engineering.

 How it works Plot the data:

The data points are plotted on a scatter plot to visualize their relationship.

Draw a line: A linear equation (a line) is drawn to represent the relationship between the data points.

Calculate errors: For each data point, the "error" or "residual" is the vertical distance from the point to the line. It is calculated as the observed value (\(y\)) minus the predicted value (\(\^{y}\)).

Square the errors: To avoid positive and negative errors canceling each other out and to penalize larger errors more heavily, each error is squared.

Sum the squared errors: All of the squared errors are added together to get a single value.Find the minimum: The method of least squares finds the specific line that makes this sum of squared errors as small as possible, or "least".

Why it's important Best fit:

It provides the most accurate linear model for a dataset by finding the line that best fits the data.

Prediction: Once the line is determined, it can be used to predict the dependent variable for new, independent variable values.

Minimizes error: The primary goal is to minimize the overall error of the model, which leads to a more reliable prediction

Ques 6 : What is Logistic Regression? How does it differ from Linear Regression?

Answer : Logistic regression predicts a categorical outcome, like spam or not-spam, by outputting a probability via a sigmoid curve, while linear regression predicts a continuous outcome, like house price, using a straight line.
 The key differences are their purpose (classification vs. regression), output (probability vs. continuous value), and the underlying mathematical model (sigmoid function vs. straight line).

 - Logistic Regression  :   

  Purpose:
  Classification problems, especially binary outcomes (e.g., yes/no, true/false).
  
   Output:
   A probability between 0 and 1, which can be used to classify an outcome. For example, a probability of \(0.8\) could mean an email is \(80\%\) likely to be spam.

   Model: Uses a sigmoid (or logistic) function to create an S-shaped curve that maps any input to a value between 0 and 1.
  
   Equation: The output is a probability value and is based on a logarithmic equation.
  
   Example: Predicting if a customer will click an ad (yes/no) or if a tumor is malignant or benign.
   
   - Linear Regression :   
   
   Purpose: Regression problems to predict a continuous value.

   Output: A continuous numerical value (e.g., $250,000 for a house price).
  
   Model: Uses a straight line to show the linear relationship between independent and dependent variables.
   
   Equation: Uses a simple linear equation, like \(y=mx+b\).
   Example: Predicting a student's test score based on hours studied, or forecasting a company's sales based on marketing spend

Quees 7 : Name and briefly describe three common evaluation metrics for regression
models.


Answer : Three common regression metrics are Mean Absolute Error (MAE), which measures the average absolute difference between predicted and actual values; Mean Squared Error (MSE), which averages the squared differences to penalize larger errors more heavily; and R-squared (\(R^{2}\)), which indicates the proportion of the variance in the dependent variable that is predictable from the independent variable(s).

- Mean Absolute Error (MAE):

The average of the absolute differences between the predicted and actual values. It is easy to interpret because it is the average magnitude of the errors in the same units as the target variable.

- Mean Squared Error (MSE):

 The average of the squared differences between the predicted and actual values. This metric heavily penalizes larger errors compared to MAE.

 - R-squared (\(R^{2}\)):

 This value represents the proportion of the variance for a dependent variable that's predictable from the independent variables. An \(R^{2}\) of \(0.8\) means that \(80\%\) of the variance in the dependent variable can be explained by the model's predictors

Ques 8 : What is the purpose of the R-squared metric in regression analysis?

Answer : The purpose of the R-squared metric in regression analysis is to indicate how well the independent variables explain the variance in the dependent variable. It measures the proportion of the variability in the dependent variable that is predictable from the independent variables. This provides a measure of the goodness of fit, with values ranging from \(0\) to \(1\) (or \(0\%\) to \(100\%\)), where a higher R-squared means the model's predictions align more closely with the actual data.

- How R-squared works:

1- Measures fit:

 R-squared is a key "goodness-of-fit" statistic that tells you how well your model's regression line or curve fits the observed data.

 2- Defines explanatory power:

  A higher R-squared indicates that a larger percentage of the variance in the target variable is explained by the model. For example, an R-squared of \(0.75\) means that \(75\%\) of the variability in the dependent variable is accounted for by the independent variables.
  
 3- Ranges from 0 to 1:\(0\):

  The model explains none of the variability in the dependent variable.\(1\): The model perfectly explains all the variability in the dependent variable.
  
  4- Simplifies interpretation:
  
   R-squared provides a single, easy-to-understand metric (often expressed as a percentage) to evaluate how strong the relationship is between the model and the data

Ques 9 : Write Python code to fit a simple linear regression model using scikit-learn
and print the slope and intercept.


```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data (replace with your actual data)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Independent variable (features)
y = np.array([2, 4, 5, 4, 5])  # Dependent variable (target)

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Print the slope (coefficient) and intercept
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

Answer : In a simple linear regression model, the equation is typically represented as:

1- Intercept ((\beta_0)):

 This is the predicted value of the dependent variable ((y)) when the independent variable ((x)) is zero. It's the point where the regression line crosses the y-axis. However, interpreting the intercept only makes sense if (x=0) is a meaningful value in your dataset.


2- Slope ((\beta_1)):

 This is the most important coefficient to interpret. It represents the change in the dependent variable ((y)) for every one-unit increase in the independent variable ((x)), assuming all other variables (in the case of multiple regression) are held constant.


For example, if you're predicting house prices ((y)) based on the size of the house in square feet ((x)), a slope ((\beta_1)) of 100 would mean that for every additional square foot of size, the predicted house price increases by $100.
